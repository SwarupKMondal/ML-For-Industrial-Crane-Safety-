```python
data = pd.read_csv("DP.csv")
```

```python
X= data[['Hs','Ht','Hd','Hr','Cv','FF','FD','SD','SA','FSR','RT','C_E','P_E','TCE','tc','fn']]
y = data['Workload']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3)
```

```python
knn5 = KNeighborsClassifier(n_neighbors = 5)
```

```python
knn5.fit(X_train, y_train)

y_pred_20 = knn5.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
print("Accuracy with k=5", accuracy_score(y_test, y_pred_20)*100)
```
